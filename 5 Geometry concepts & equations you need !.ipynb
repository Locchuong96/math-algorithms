### 1/ Euclid distance



```python
import math 
def distance(point1,point2):
    sum = 0
    for i in range(len(point1)):
        sum += (point1[i] + point2[i])**2
    return math.sqrt(sum)

point1 = (0,0)
point2 = (3,4)

distance(point1,point2)
```

### 2/ Orthogonal projection



```python
def projection_point2line(point,m,b):
    x,y = point
    m2 = -1/m
    b2 = y - m2*x
    x_intersection = -(b-b2)/(m-m2)
    y_intersection = m2 * x_intersection + b2
    return (x_intersection, y_intersection)

projection_point2line((2,3),-3/5,6)
```

### 4/ Interpolation



```python
def interpolation(point1,point2,alpha):
    x1,y1 = point1
    x2,y2 = point2 
    x = x2 * alpha + x1 * (1 -alpha)
    y = y2 * alpha + y1 * (1- alpha)
    return x,y
```
